```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
dataset = pd.read_csv('v4.csv')

# for i in range
# dataset.iloc[: ,  ]
dataset.rename(columns={'Total Living Area': 'Total_Living_Area'}, inplace=True)
dataset
```

```python
dataset['Total_Living_Area'] = dataset['Total_Living_Area'].astype('float64')
dataset['Deed_Date'] = dataset['Deed_Date'].astype('datetime64[ns]')
dataset['Council_District'] = dataset['Council_District'].astype('string')
dataset['Police_District'] = dataset['Police_District'].astype('string')
dataset['Neighborhood'] = dataset['Neighborhood'].astype('string')
dataset['Print_Key'] = dataset['Print_Key'].astype('string')
dataset['Street'] = dataset['Street'].astype('string')
dataset['Front'] = dataset['Front'].astype('float64')
dataset['Land_Value'] = dataset['Land_Value'].astype('float64')
dataset['Total_Value'] = dataset['Total_Value'].astype('float64')
dataset['Sale_Price'] = dataset['Sale_Price'].astype('float64')
dataset['Year_Built'] = dataset['Year_Built'].astype('int64')
dataset['Census_Tract'] = dataset['Census_Tract'].astype('float64')
dataset.dtypes
```

```python
x=[]
for i in range (dataset['Deed_Date'].size):
  x.append(int(dataset['Deed_Date'].iloc[i].year))
dataset['Deed_Date']=x
```

```python
fig, axs = plt.subplots(2,3, figsize = (10,5))
plt1 = sns.boxplot(dataset['Front'], ax = axs[0,0])
plt2 = sns.boxplot(dataset['Depth'], ax = axs[0,1])
plt3 = sns.boxplot(dataset['Deed_Date'], ax = axs[0,2])
plt1 = sns.boxplot(dataset['Sale_Price'], ax = axs[1,0])
plt2 = sns.boxplot(dataset['Sft_Price'], ax = axs[1,1])
plt3 = sns.boxplot(dataset['Total_Living_Area'], ax = axs[1,2])

plt.tight_layout()
```

```python

print(dataset.Depth.size)
# outlier treatment for Front
Q1 = dataset.Front.quantile(0.25)
Q3 = dataset.Front.quantile(0.75)
IQR = Q3 - Q1
dataset = dataset[(dataset.Front >= Q1 - 1.5*IQR) & (dataset.Front <= Q3 + 1.5*IQR)]
plt.boxplot(dataset.Front)

# outlier treatment for Depth
Q1 = dataset.Depth.quantile(0.25)
Q3 = dataset.Depth.quantile(0.75)
IQR = Q3 - Q1
dataset = dataset[(dataset.Depth >= Q1 - 1.5*IQR) & (dataset.Depth <= Q3 + 1.5*IQR)]
plt.boxplot(dataset.Depth)

# outlier treatment for Sale_Price
Q1 = dataset.Sale_Price.quantile(0.25)
Q3 = dataset.Sale_Price.quantile(0.75)
IQR = Q3 - Q1
dataset = dataset[(dataset.Sale_Price >= Q1 - 1.5*IQR) & (dataset.Sale_Price <= Q3 + 1.5*IQR)]
plt.boxplot(dataset.Sale_Price)

# outlier treatment for Sft_Price
Q1 = dataset.Sft_Price.quantile(0.25)
Q3 = dataset.Sft_Price.quantile(0.75)
IQR = Q3 - Q1
dataset = dataset[(dataset.Sft_Price >= Q1 - 1.5*IQR) & (dataset.Sft_Price <= Q3 + 1.5*IQR)]
plt.boxplot(dataset.Sft_Price)

# outlier treatment for Total_Living
Q1 = dataset.Total_Living_Area.quantile(0.25)
Q3 = dataset.Total_Living_Area.quantile(0.75)
IQR = Q3 - Q1
dataset = dataset[(dataset.Total_Living_Area >= Q1 - 1.5*IQR) & (dataset.Total_Living_Area <= Q3 + 1.5*IQR)]
plt.boxplot(dataset.Total_Living_Area)

dataset.Depth.size
```

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(dataset , train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
num_vars = ['Front', 'Depth', 'Sale_Price', 'Sft_Price', 'Beds','Total_Living_Area','Zipcode']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
plt.figure(figsize = (16, 10))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

```python
y_train = df_train.pop('Sft_Price')
X_train = df_train
y_train
```

```python
X_train.pop('Print_Key')
X_train.pop('Street')
X_train.pop('Council_District')
X_train.pop('Police_District')
X_train.pop('Neighborhood')
# X_train.pop('Deed_Date')
X_train.pop('Location')
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
lm
```

```python
rfe = RFE(lm, step = 20)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable
import statsmodels.api as sm
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```

```python
#Let's see the summary of our linear model
print(lm.summary())
```

```python
# Calculate the VIFs for the model
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
y_train_price = lm.predict(X_train_rfe)
```

```python
res = (y_train_price - y_train)
```
